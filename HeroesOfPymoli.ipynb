## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Pop_Group= df.groupby(["Item Name"])
Pop_Comparison=Pop_Group.count()
Popular_Items= Pop_Comparison.sort_values("Item ID", ascending=False)
Pop_Comparison.head()
```

```python
# Create new table
Pop_type = df[["Item ID", "Item Name", "Price",]]
#Pop_Item_group = Pop_type.groupby(["Item ID"]).size()
#Pop_Item_group
#pd.merge(Pop_type,Pop_Item_group,on='Item ID')
#Pop_type.join(Pop_Item_group, on='Item ID')
Pop_type['count'] = Pop_type.groupby('Item ID')['Item ID'].transform('count')
Pop_type=Pop_type.sort_values(by=['count'], ascending=False)
Pop_type.head(30)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#df.groupby(['Item ID', 'Item Name'])['Price'].agg('sum')
Pop_type['Real Total Value'] = Pop_type.groupby('Item ID')['Price'].agg('sum')
Pop_type=Pop_type.sort_values(by=['Total Value'], ascending=False)
Pop_type.head(30)
```
